```python
faqs = """
গরু আমাদের জীবনে অত্যন্ত গুরুত্বপূর্ণ একটি প্রাণী। এটি কৃষি কাজের জন্য অত্যন্ত উপযোগী এবং দুধ উৎপাদনের প্রধান উৎস। গরুর দুধ থেকে তৈরি করা হয় দই, ঘি, মাখন এবং আরও অনেক পুষ্টিকর খাদ্য। গ্রামীণ অর্থনীতিতে গরুর অবদান অপরিসীম, কারণ এটি কৃষকের প্রধান সহায়ক হিসেবে কাজ করে। তাছাড়া, গরু অনেক ধর্মীয় ও সাংস্কৃতিক আচার-অনুষ্ঠানেরও একটি গুরুত্বপূর্ণ অংশ।

গ্রামে সাধারণত গরুকে পরিবারের সদস্যের মতো যত্ন নেওয়া হয়। গরুর চারণভূমিতে খাওয়ার দৃশ্য গ্রামীণ পরিবেশের একটি চিরচেনা দৃশ্য।

গরুর মাংস অনেক অঞ্চলে একটি প্রধান খাদ্য হিসেবে ব্যবহৃত হয়, যদিও বিভিন্ন সংস্কৃতির মধ্যে এই বিষয়ে ভিন্নতা রয়েছে। গরুর চামড়া থেকে তৈরি হয় বিভিন্ন ধরনের চামড়াজাত পণ্য, যা শিল্পের একটি বড় অংশ জুড়ে রয়েছে।

এভাবে গরু আমাদের অর্থনৈতিক, সাংস্কৃতিক ও সামাজিক জীবনে একটি গুরুত্বপূর্ণ ভূমিকা পালন করে চলেছে।"""
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([faqs])
```

```python
tokenizer.word_index
```

```python
input_sequences = []
for sentence in faqs.split('\n'):
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

  for i in range(1,len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences
```

```python

max_len = max([len(x) for x in input_sequences])
```

```python
print(max_len)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences,maxlen = max_len,padding = 'pre')
```

```python
padded_input_sequences
```

```python
X = padded_input_sequences[:,:-1]
y = padded_input_sequences[:,-1]
```

```python
X.shape
```

```python
y.shape
```

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(y,num_classes = 90)
#num_classes  =  unique_word + 1
```

```python
y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense
```

```python
model = Sequential()
model.add(Embedding(90, 100, input_length=55))
model.add(LSTM(150))

model.add(Dense(90, activation='softmax'))
#number of unique word = 89
#maxlength = 56
#input_length = 55
```

```python
model.compile(loss = 'categorical_crossentropy', optimizer = 'adam',metrics = ['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X,y,epochs=100)
```

```python
import numpy as np
import time
text = "গরু আমাদের জীবনে অত্যন্ত গুরুত্বপূর্ণ একটি প্রাণী"

for i in range(10):
  # tokenize
  token_text = tokenizer.texts_to_sequences([text])[0]
  # padding
  padded_token_text = pad_sequences([token_text], maxlen=22, padding='pre')
  # predict
  pos = np.argmax(model.predict(padded_token_text))

  for word,index in tokenizer.word_index.items():
    if index == pos:
      text = text + " " + word
      print(text)
      time.sleep(2)
```
